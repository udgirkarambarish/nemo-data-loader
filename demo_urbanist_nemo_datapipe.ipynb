```python
import torch
from torch.utils.data import DataLoader

from urbanist_nemo_loader import UrbanistNeMoDataPipe
```

# Expected directory structure:
#
# data/
# ├── dem.tif
# └── HRF.tif


```python
datapipe = UrbanistNeMoDataPipe(
    data_dir="data/",
    dem_range=(0.0, 600.0),
    hrf_range=(0.0, 6.0),
)
```

```python
loader = DataLoader(
    datapipe,
    batch_size=1
)
```

```python
batch = next(iter(loader))

features = batch["features"]
labels = batch["labels"]

features, labels
```

```python
print("Features shape:", features.shape)
print("Labels shape:", labels.shape)
```
